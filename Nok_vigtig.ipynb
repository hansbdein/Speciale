```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

T0=2
T=np.logspace(np.log(T0),np.log(0.01),base=np.exp(1),num=1000)
Q=1.293         #neutron proton mass difference in MeV

def eqD(T):    
    return np.exp(Q/T)

def Xn_eq(T):
    return 1/(eqD(T)+1)

fig=plt.figure()




plt.plot(T,2*Xn_eq(T),'r-')

n_life=886.7    #neutron lifetime in seconds

def np_rate(x):
    return 255/(n_life*x**5)*(12+6*x+x**2)

H=1.13          #hubble constant at Q=T in seconds^-1

# Define the differential equation
def dodel_3_27(X,x):
    dXndx=x*np_rate(x)/H*(np.exp(-x)-X*(1+np.exp(-x)))
    return dXndx

# Define the initial condition
X0 = Xn_eq(T0)

# Define the time points to solve the differential equation
x = Q/T

print(dodel_3_27(2,Q/T0))

# Solve the differential equation
Xn = odeint(dodel_3_27, X0, x)

# Plot the results
plt.plot(T,2*Xn, 'r--', label='without equilibrium')
plt.xlabel('Temperature (MeV)')
plt.ylabel('abundance')
plt.legend()

'''
'''

plt.ylim(3e-5, 1)
plt.yscale('log')
plt.xlim(2, 0.02)
plt.xscale('log')
plt.show()
```
